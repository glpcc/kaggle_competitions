```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/train.csv')
```

```python
df['Embarked'].fillna('N',inplace=True)
test = {
    'N':0,
    'S':1,
    'C':2,
    'Q':3
}
df['Embarked'] = df['Embarked'].map(lambda k: test[k])
```

```python

indexes = {
    'C':0,
    'B':1,
    'D':2,
    'E':3,
    'A':4,
    'F':5,
    'G':6,
    'T':7
}

def get_section_code(k: str,avrgs = []):
    num_cabins = {
        'C':0,
        'B':0,
        'D':0,
        'E':0,
        'A':0,
        'F':0,
        'G':0,
        'T':0
    }
    result = [0,0,0,0,0,0,0,0]
    t = ord('N')
    k = str(k)
    if k == 'nan':
        return result
    sections = set()
    i = 0
    while i < len(k):
        if k[i] != ' ' and not k[i].isdigit():
            j = i
            while j < len(k) and k[j] != ' ':
                j += 1
            if k[i+1:j] != '':
                result[indexes[k[i]]] += int(k[i+1:j])
                num_cabins[k[i]] += 1
            elif len(avrgs) != 0:
                result[indexes[k[i]]] += avrgs[indexes[k[i]]]
                num_cabins[k[i]] += 1
        i += 1 
    for i in num_cabins:
        if num_cabins[i] > 0:
            result[indexes[i]] /= num_cabins[i]
            
    return result

#df[df['Cabin'].map(check_if_all_same_section)]['PassengerId']
df['Section'] = df['Cabin'].map(get_section_code)
sections = np.array(df['Section'].to_list())
avrgs = sections.mean(axis=0,where=sections!=0)
# Assing T section avrg to arbitrary num as there is a lack of data in that section
avrgs[-1] = 30
df['Section'] = df['Cabin'].map(lambda k: get_section_code(k,avrgs))


```

```python
for i in indexes:
    df[f'Section{i} Value'] = df['Section'].map(lambda k: k[indexes[i]])

for i in indexes:
    df[f'Section{i} Value'] = (df[f'Section{i} Value']-df[f'Section{i} Value'].min())/(df[f'Section{i} Value'].max()-df[f'Section{i} Value'].min()) 

df.drop(columns=['Section','Cabin']);
```

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
df['Age'] = (df['Age'] - df['Age'].min())/(df['Age'].max()-df['Age'].min())
df['SexVal'] = df['Sex'].map(lambda k: 1 if k == 'female' else 0)
df['Fare'] = (df['Fare']-df['Fare'].min())/(df['Fare'].max()-df['Fare'].min())
```

```python
df.drop(columns=['Sex','Name','Ticket','Section','Cabin'],inplace=True)
```
