```python
import pandas as pd
import numpy as np
from matplotlib import pyplot
```

```python
process_train_csv = True
if process_train_csv:
    df = pd.read_csv('data/train_processed.csv')
else:
    df = pd.read_csv('data/test_processed.csv')
```

```python
df_num = df.select_dtypes(include = ['float64', 'int64'])
df_num.drop('MSSubClass',inplace=True,axis=1)
# df_num.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8);
```

```python
# Standarize The Values
ids = df_num["Id"]
if process_train_csv:
    saleprice = df_num["SalePrice"]
std_df = (df_num-df_num.mean())/df_num.std()
if process_train_csv:
    std_df["SalePrice"] = saleprice
std_df["Id"] = ids
```

```python
# Save the csvs
if process_train_csv:
    std_df.to_csv('data/num_train_processed.csv',index=False)
else:
    std_df.to_csv('data/num_test_processed.csv',index=False)
```
